```python
import numpy as np
```

```python
def split_data(data, ratio):
    np.random.seed(42)
    shuffled = np.random.permutation(len(data))
    test_set_size = int(len(data) * ratio)
    test_indicies = shuffled[:test_set_size]
    train_indicies = shuffled[test_set_size:]
    return data.iloc[train_indicies], data.iloc[test_indicies]
```

```python
train, test = split_data(housing, 0.2)
```

```python
X_train = train[['Area', 'Hall', 'Bathroom', 'Bed room', 'Kitchen', 'Terrace']].to_numpy()
X_test = test[['Area', 'Hall', 'Bathroom', 'Bed room', 'Kitchen', 'Terrace']].to_numpy()
```
